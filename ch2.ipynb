```python
import networkx as nx
%matplotlib inline
G=nx.Graph()
nodes_to_add=['a','b','c','d']
G.add_nodes_from(nodes_to_add)
edges_to_add=[('a', 'b'),('a', 'c'), ('b', 'c'), ('c', 'd')]
G.add_edges_from(edges_to_add)
nx.draw(G,with_labels=True, font_color= 'red')
```

```python
nx.draw(G,
        with_labels=True,
        node_color='red',
        node_size=1000,
        font_color='blue',
        font_size=16,
        )
```

```python
G = nx.Graph()

G.add_nodes_from(['lion','tiger','cheta',13])

G.add_edge('cat','virus')

nx.draw(G, with_labels=True, font_color='white', node_size=1000)
```
